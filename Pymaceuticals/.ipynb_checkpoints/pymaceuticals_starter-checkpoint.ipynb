```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
import random

# Study data files - I chanted this to reflect the file directory
mouse_metadata_path = "../data/Mouse_metadata.csv"
study_results_path = "../data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)


#View the data
print(mouse_metadata)
print(study_results)
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary dataframe. 
treatment_groups = clean_df.groupby(['Drug Regimen'])

mean_tumor = treatment_groups['Tumor Volume (mm3)'].mean().to_frame('Mean')
median_tumor = treatment_groups['Tumor Volume (mm3)'].median().to_frame('Median') 
var_tumor = treatment_groups['Tumor Volume (mm3)'].var().to_frame('Var')
std_tumor = treatment_groups['Tumor Volume (mm3)'].std().to_frame('Std')
sem_tumor = treatment_groups['Tumor Volume (mm3)'].sem().to_frame('SEM') 

#merge the data
summary_df = mean_tumor.merge(median_tumor,on='Drug Regimen').merge(var_tumor,on='Drug Regimen').merge(std_tumor,on='Drug Regimen').merge(sem_tumor,on='Drug Regimen')

print(summary_df)

```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Using the aggregation method, produce the same summary statistics in a single line

summary_df = treatment_groups['Tumor Volume (mm3)'].agg(['mean', 'median', 'var', 'std', 'sem'])

print(summary_df)
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

#unique_mice is defined above - the last row for each mouse

treatments = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

#create a filter to include only identified treatments
trtmt_filt = unq_mice['Drug Regimen'].isin(treatments)

#apply filter to unique mice
end_vals = unq_mice[trtmt_filt].sort_values(by='Drug Regimen')
end_vals
```

```python
# Put treatments into a list for for loop (and later for plot labels)
# 'end_vals' defined in previous step = data for last tumor volume measurement of each mouse

# Create empty list to fill with tumor vol data (for plotting)
#tmr_vol_data = []

trial_bnds_df = pd.DataFrame(columns=['Treatment','Lower Quartile','Median','Upper Quartile','Inter Quartile Range','Lower Bound','Upper Bound'])



# Calculate the IQR and quantitatively determine if there are any potential outliers. 

for treatment in treatments:
    
    # Locate the rows which contain mice on each drug and get the tumor volumes
    tgt_grp = end_vals[end_vals['Drug Regimen'] == treatment]
    
    # Determine outliers using upper and lower bounds
    quartiles = tgt_grp['Tumor Volume (mm3)'].quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    mdn = quartiles[0.5]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    
    #push bounds into dataframe
    trial_bnds_df.loc[len(trial_bnds_df.index)] = [treatment, lowerq, mdn, upperq, iqr, lower_bound, upper_bound] 

#sanity check
trial_bnds_df

    
```

```python
#Check for outliers in each treatment group
outliers_df = pd.DataFrame()

for treatment in treatments:
    #get the bounds from the table above
    lower_bound =trial_bnds_df.loc[trial_bnds_df['Treatment'] == treatment, 'Lower Bound'].values[0]

    upper_bound =trial_bnds_df.loc[trial_bnds_df['Treatment'] == treatment, 'Upper Bound'].values[0]

    #search for anything outside of the bounds
    tgt_grp_2 = end_vals.loc[(end_vals['Drug Regimen'] == treatment) & ((end_vals['Tumor Volume (mm3)'] < lower_bound) | (end_vals['Tumor Volume (mm3)'] > upper_bound))]
    
    #put the outliers in outliers_df outside of loop
    outliers_df = outliers_df.append(tgt_grp_2)

#see the results
outliers_df
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
#Data is in end_vals
#Bounds are in trial_bnds_df

#set up the data in series 

infubinol_vols = end_vals.loc[end_vals['Drug Regimen'] == "Infubinol", 'Tumor Volume (mm3)']
capomulin_vols = end_vals.loc[end_vals['Drug Regimen'] == "Capomulin", 'Tumor Volume (mm3)']
ramicane_vols = end_vals.loc[end_vals['Drug Regimen'] == "Ramicane", 'Tumor Volume (mm3)']
ceftamin_vols = end_vals.loc[end_vals['Drug Regimen'] == "Ceftamin", 'Tumor Volume (mm3)']


blue_circle = dict(markerfacecolor='b', marker='o')
data = [infubinol_vols, capomulin_vols, ramicane_vols, ceftamin_vols ]
fig, ax = plt.subplots()
ax.set_title('Mouse Trial Results')
ax.set_ylabel('Tumor Volume (mm3)')
ax.boxplot(data, flierprops=blue_circle)
plt.xticks([1, 2, 3, 4], ['Infubinol', 'Capomulin', 'Ramicane', 'Ceftamin'])

plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin
#clean_df includes all data - duplicates
#choose a mouse id
capomulin_mice = clean_df.loc[clean_df['Drug Regimen'] == 'Capomulin', 'Mouse ID']
np.random.seed(10)
mickey = np.random.choice(capomulin_mice)

#return mickey's data
mick_df = clean_df.loc[clean_df['Mouse ID'] == mickey]

#plot mickey's data
mick_df.plot(x='Timepoint', y='Tumor Volume (mm3)', title='Capomulin Mouse Tumor over Time', label=mickey)
plt.ylabel('Tumor Volume (mm3)')
plt.show()
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
#return the Capomulin Data Set from clean_df
capomulin_df = clean_df.loc[clean_df['Drug Regimen'] == 'Capomulin']


#group and take mean / make into a dataframe /because data is in index, reset it first
wgt_tmr = capomulin_df.groupby('Weight (g)')['Tumor Volume (mm3)'].mean().to_frame()
wgt_tmr = wgt_tmr.reset_index()

#scatter plot the df - 
wgt_tmr.plot(kind='scatter', x='Weight (g)', y='Tumor Volume (mm3)', title='Average Tumor Volume by Mouse Weight')
plt.ylabel('Average Tumor Volume (mm3)')
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

#get data
x_values = wgt_tmr['Weight (g)']
y_values = wgt_tmr['Tumor Volume (mm3)']

#regression line
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

#scatter plot
plt.scatter(x_values,y_values)
plt.title('Mouse Weight vs. Tumor Volume')
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(20,40),fontsize=15,color="red")
plt.xlabel('Mouse Weight (g)')
plt.ylabel('Avg Tumor Volume (mm3)')

#display data
print(f"The correlation coefficient between weight and tumor volume is {round(st.pearsonr(x_values,y_values)[0],2)}.")
print(f"The r-squared is: {rvalue**2}")
plt.show()
```
