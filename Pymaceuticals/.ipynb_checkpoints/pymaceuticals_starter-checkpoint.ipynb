## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files - I chanted this to reflect the file directory
mouse_metadata_path = "../data/Mouse_metadata.csv"
study_results_path = "../data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)


#View the data
print(mouse_metadata)
print(study_results)
```

```python
# Combine the data into a single dataset (Mouse ID)
combined_data = mouse_metadata.merge(study_results, how='outer', on='Mouse ID')

# Display the data table for preview
print(combined_data)
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID. 
# I didn't drop the mouse - I just dropped the duplicate records.  I wasn't sure what was intended here.  I could have eliminated the mouse entirely, but that didn't seem to make sense, so I just dropped duplated rows, keeping one row for each mouse/time point.
clean_df = combined_data.drop_duplicates(subset=['Mouse ID', 'Timepoint'], keep='last')
clean_df
```

```python
# Checking the number of mice in the clean DataFrame.
clean_df['Mouse ID'].count()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary dataframe. 
treatment_groups = clean_df.groupby(['Drug Regimen'])

mean_tumor = treatment_groups['Tumor Volume (mm3)'].mean().to_frame('Mean')
median_tumor = treatment_groups['Tumor Volume (mm3)'].median().to_frame('Median') 
var_tumor = treatment_groups['Tumor Volume (mm3)'].var().to_frame('Var')
std_tumor = treatment_groups['Tumor Volume (mm3)'].std().to_frame('Std')
sem_tumor = treatment_groups['Tumor Volume (mm3)'].sem().to_frame('SEM') 

#merge the data
summary_df = mean_tumor.merge(median_tumor,on='Drug Regimen').merge(var_tumor,on='Drug Regimen').merge(std_tumor,on='Drug Regimen').merge(sem_tumor,on='Drug Regimen')

print(summary_df)

```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Using the aggregation method, produce the same summary statistics in a single line

summary_df = treatment_groups['Tumor Volume (mm3)'].agg(['mean', 'median', 'var', 'std', 'sem'])

print(summary_df)
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

#unique_mice is defined above - the last row for each mouse

treatments = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

#create a filter to include only identified treatments
trtmt_filt = unq_mice['Drug Regimen'].isin(treatments)

#apply filter to unique mice
end_vals = unq_mice[trtmt_filt].sort_values(by='Drug Regimen')
end_vals
```

```python
# Put treatments into a list for for loop (and later for plot labels)
# 'end_vals' defined in previous step = data for last tumor volume measurement of each mouse

# Create empty list to fill with tumor vol data (for plotting)
#tmr_vol_data = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 

for treatment in treatments:
    
    # Locate the rows which contain mice on each drug and get the tumor volumes
    tgt_grp = end_vals[end_vals['Drug Regimen'] == treatment]
    
    # Determine outliers using upper and lower bounds
    quartiles = tgt_grp['Tumor Volume (mm3)'].quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    
    print(f"{treatment} Summary: ________________________________________________")
    print(f"The lower quartile of tumor volume is: {lowerq}")
    print(f"The upper quartile of tumor volume is: {upperq}")
    print(f"The interquartile range of tumor volume is: {iqr}")
    print(f"The the median of tumor volume is: {quartiles[0.5]} ")

    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    print(f"Values below {lower_bound} could be outliers.")
    print(f"Values above {upper_bound} could be outliers.")
    
    

    
```
