## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files - I chanted this to reflect the file directory
mouse_metadata_path = "../data/Mouse_metadata.csv"
study_results_path = "../data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)


#View the data
print(mouse_metadata)
print(study_results)
```

```python
# Combine the data into a single dataset (Mouse ID)
combined_data = mouse_metadata.merge(study_results, how='outer', on='Mouse ID')

# Display the data table for preview
print(combined_data)
```

```python
# Checking the number of mice.
combined_data['Mouse ID'].count()
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
combined_data['Mouse ID'].value_counts()
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicates = combined_data[combined_data.duplicated(['Mouse ID', 'Timepoint'])]

#get the duplicated ids
dup_ids = duplicates['Mouse ID']
dup_ids = combined_data['Mouse ID'].isin(dup_ids)

#select corresponding rows from combined import data.
filtered_df = combined_data[dup_ids]
filtered_df
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID. 
# I didn't drop the mouse - I just dropped the duplicate records.  I wasn't sure what was intended here.combined_data
cleaned_data = combined_data.drop_duplicates(subset=['Mouse ID', 'Timepoint'], keep='last')
print(cleaned_data)
```

```python
# Checking the number of mice in the clean DataFrame.
cleaned_data['Mouse ID'].count()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary dataframe. 
drug_groups = cleaned_data.groupby(['Drug Regimen'])

mean_tumor = drug_groups['Tumor Volume (mm3)'].mean().to_frame('Mean')
median_tumor = drug_groups['Tumor Volume (mm3)'].median().to_frame('Median') 
var_tumor = drug_groups['Tumor Volume (mm3)'].var().to_frame('Var')
std_tumor = drug_groups['Tumor Volume (mm3)'].std().to_frame('Std')
sem_tumor = drug_groups['Tumor Volume (mm3)'].sem().to_frame('SEM') 

#merge the data
summary_df = mean_tumor.merge(median_tumor,on='Drug Regimen').merge(var_tumor,on='Drug Regimen').merge(std_tumor,on='Drug Regimen').merge(sem_tumor,on='Drug Regimen')

print(summary_df)

```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Using the aggregation method, produce the same summary statistics in a single line
drug_groups = cleaned_data.groupby(['Drug Regimen'])
summary_df = drug_groups['Tumor Volume (mm3)'].agg(['mean', 'median', 'var', 'std', 'sem'])

print(summary_df)
```

## Bar and Pie Charts

```python
# Generate a bar plot showing the total number of unique mice tested on each drug regimen using pandas.
mouse_count = drug_groups['Mouse ID'].nunique().to_frame()
mouse_count.plot(kind='bar', legend=False, title='Unique Mice per Treatment', ylabel='# of Unique Mice')
plt.show()
```

```python
# Generate a bar plot showing the total number of unique mice tested on each drug regimen using pyplot.

plt.bar(mouse_count.index, mouse_count['Mouse ID'], width=.6)
plt.title('Unique Mice per Treatment')
plt.xlabel('Drug Regimen')
plt.ylabel('# of Unique Mice')
plt.xticks(rotation=90)
plt.show()
```

```python
# Generate a pie plot showing the distribution of female versus male mice using pandas
# plt.show()
#create df with only uniqe mice
unique_mice = cleaned_data.drop_duplicates(['Mouse ID'], keep='last')
unique_mice['Sex'].value_counts().plot.pie(y='Sex', title='Study Mice by Sex', autopct='%1.0f%%')
plt.axis("equal")
plt.show()
```

```python
# Generate a pie plot showing the distribution of female versus male mice using pyplot
sex_counts = unique_mice['Sex'].value_counts()
labels = sex_counts.keys()
plt.pie(sex_counts, labels=labels, autopct='%1.0f%%')
plt.title('Study Mice by Sex')
plt.ylabel('Sex')
plt.axis("equal")
plt.show()

```
